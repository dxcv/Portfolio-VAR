```python
import pandas as pd
import numpy as np 

LN_Port_5 = pd.read_csv('Port5_LN.csv')

LN_Port_5.head()
```

```python
covariance = LN_Port_5.cov()


chol = np.linalg.cholesky(covariance)
```

```python
to_do = [0,1,2]
n_sims = 3000
rand_numbers = pd.DataFrame()

for x in to_do:
    rand_numbers[x] = np.random.normal(0, 1, n_sims)
    
sim = []
outtest = []
for x in list(range(0, n_sims)):
    outtest = chol.dot(np.array(rand_numbers.loc[x, :]))
    sim.append(outtest)
    
sim_returns = pd.DataFrame(sim)

sim_price = (1 + sim_returns*100) * [1.0335, 0.9978, 0.9044]
```
