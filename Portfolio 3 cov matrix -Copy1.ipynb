```python
# RISK FACTORS 
'''for the portolfio... our factors will be:
underlying currencies --> EUR, USD, JPY, CHF
Forward AUD zero curves at maturities --> 123, 247, 276, 374
Forein Zero curves --> USD(123), EUR(247), JPY(276), CHF(374) 
12 RISK FACTORS 
'''

import pandas as pd 
import numpy as np 
import scipy.interpolate
```

```python
hist_bond_yields = pd.read_csv('hist_bond_yields.csv')
#store array X of given yield maurities in days 
X = [0,30.41666667,60.83333333,91.25,182.5,273.75,365,730,1095,1460,1825,2190,2555,2920,3285]

#Array of yields needed to be interpreted 
tointerp = [123, 247, 276, 374]

df = []

#to loop through hist_bond_yields 
Dateindex = list(range(0,2762))

#interp function 
for row in Dateindex:
    
    #this is interpolation function that iterates over the bond_yields data frame row by row looking at yields 

    interp = scipy.interpolate.interp1d(X, hist_bond_yields.iloc[row, 1:], bounds_error=False, fill_value=scipy.nan)
    
    outtest = []
    
    for intx in tointerp:
        
        outtest.append(float(interp(intx)))
        
    df.append(outtest)
    outtest = []
    
interp_yields = pd.DataFrame(df, columns = tointerp)

```

```python
EUR_ZC = pd.read_csv('EUR_ZC.csv')
JPY_ZC = pd.read_csv('JPY_ZC.csv')
US_ZC = pd.read_csv('US_ZC.csv')
Swiss_ZC = pd.read_csv('Swiss_ZC.csv')
```

```python
EUR_ZC.head()
```

```python
EUR_interp = []
for row in Dateindex:
    interp = scipy.interpolate.interp1d([182.5, 273.75], EUR_ZC.iloc[row, :], bounds_error=False, fill_value=scipy.nan)
    EUR_interp.append(float(interp(247)))
    
JPY_interp = []
for row in Dateindex:
    interp = scipy.interpolate.interp1d([273.75, 365], JPY_ZC.iloc[row, :], bounds_error=False, fill_value=scipy.nan)
    JPY_interp.append(float(interp(276)))

US_interp = []
for row in Dateindex:
    interp = scipy.interpolate.interp1d([91.25, 182.5], US_ZC.iloc[row, :], bounds_error=False, fill_value=scipy.nan)
    US_interp.append(float(interp(123)))
    
Swiss_interp = []
for row in Dateindex:
    interp = scipy.interpolate.interp1d([365, 730], Swiss_ZC.iloc[row, :], bounds_error=False, fill_value=scipy.nan)
    Swiss_interp.append(float(interp(374)))

```

```python
Foreign_interp = pd.DataFrame()
Foreign_interp['JPY_int'] = JPY_interp
Foreign_interp['EUR_int'] = EUR_interp
Foreign_interp['US_int'] = US_interp
Foreign_interp['Swiss_int'] = Swiss_interp
```

```python
risk_factors = pd.read_csv('port_3_riskfactors.csv')
risk_factors.head()
```

```python
port3_cov_matrix = risk_factors.cov()
port3_cov_matrix
```

```python
port3_corr_matrix = risk_factors.corr()
port3_corr_matrix
```

```python
cov_matrix = port3_cov_matrix

Covar_X = cov_matrix.dot(X_vector)
Covar_X
```

```python
X_Covar_X = Covar_X.dot(X_vector)
X_Covar_X
```

```python
sqrt_covar = np.sqrt(X_Covar_X)
sqrt_covar
```

```python
chol1 = np.linalg.cholesky(cov_matrix)

for r_index, row in enumerate(chol1):
    for c_index, col in enumerate(row):
        # ensures we're in the bottom half of the board
        if c_index <= len(chol1[0])/2 or r_index <= len(chol1)/2:
            chol1[r_index][c_index] = chol1[c_index][r_index]
            
#random numbers
rand_numbers = pd.DataFrame()
n_sims = 10000

for x in list(range(0,12)):
    rand_numbers[x] = np.random.normal(0, 1, n_sims)
    
sim = []
outtest = []
for x in list(range(0, n_sims)):
    outtest = chol1.dot(np.array(rand_numbers.loc[x, :]))
    sim.append(outtest)

MC_sim_returns = pd.DataFrame(sim)
MC_sim_returns.columns = port3_corr_matrix.columns
```
