```python
Bond_portfolio = pd.read_csv('Bond_mat_CF.csv')
Bond_portfolio
```

```python
w = list(i for i in range(0,21))
ZR = []
for i in w:
    mat = Bond_portfolio.iloc[ i , 0]
    value = float(interp(mat))
    #Bond_portfolio['ZR'] = Bond_portfolio.append(value)
    ZR.append(value)

Bond_portfolio['ZR'] = ZR
```

```python
Bond_portfolio
```

```python
DF = []
for i in w:
    val = (Bond_portfolio.iloc[i, 2] / 100) * (Bond_portfolio.iloc[i,0] / 365)
    DisFact = np.exp(-val)
    DF.append(DisFact)
Bond_portfolio['DF'] = DF
```

```python
PV = []
for i in w:
    PresentVal = ((Bond_portfolio.iloc[i, 1]) * (Bond_portfolio.iloc[i, 3]))
    PV.append(PresentVal)

Bond_portfolio['PV'] = PV
```
